## 1. Tầng 1: Syntax & Semantics

### 1.1. Khai Báo Biến

```python
# Mutable - Biến thường
x = 10
name = "Python"
is_active = True
prices = [1.5, 2.0, 3.5]

# Type annotation (Python 3.5+)
x: int = 10
name: str = "Python"
is_active: bool = True
```

```python
# Immutable - Hằng số
# Quy ước: viết hoa
MAX_SIZE = 100
PI = 3.14159

# Tuple cho immutable data
COORDS = (10, 20)
```

```python
# Global variable
global_var = "I am global"

def access_global():
    print(global_var)

def modify_global():
    global global_var
    global_var = "Modified"
```

### 1.2. Khai Báo Hàm

```python
# Function cơ bản
def greet():
    print("Hello!")

def add(a, b):
    return a + b

# Function với type hints
def greet(name: str) -> str:
    return f"Hello, {name}!"
```

```python
# Parameters
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

# Named arguments
def create_user(name, email, age):
    return {"name": name, "email": email, "age": age}

user = create_user(age=25, name="John", email="john@example.com")

# *args và **kwargs
def func(*args, **kwargs):
    print(args)    # Tuple
    print(kwargs)  # Dictionary

func(1, 2, 3, name="John", age=25)
```

```python
# Lambda / Arrow Function
square = lambda x: x ** 2
add = lambda a, b: a + b

# Dùng với higher-order functions
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x ** 2, numbers))
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(squared, evens)
```

```python
# Closure
def outer(x):
    def inner(y):
        return x + y
    return inner

closure = outer(10)
closure(5)  # 15
```

### 1.3. Vòng Lặp

```python
# For với range
for i in range(5):
    print(i, end=" ")  # 0 1 2 3 4

# For với list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
# While loop
count = 0
while count < 5:
    print(count, end=" ")
    count += 1
print()
```

```python
# List comprehension
squares = [x**2 for x in range(10)]
evens = [x for x in range(20) if x % 2 == 0]
print(squares)
print(evens)
```

### 1.4. Điều Kiện

```python
# If-else
age = 18
if age >= 18:
    print("Adult")
else:
    print("Minor")

# Ternary operator
status = "Adult" if age >= 18 else "Minor"
print(status)
```

```python
# Match expression (Python 3.10+)
def http_status(status):
    match status:
        case 200: return "OK"
        case 404: return "Not Found"
        case 500: return "Server Error"
        case _: return "Unknown"

print(http_status(200))
```

```python
from enum import Enum, auto

class Color(Enum):
    RED = 1
    GREEN = 2
    BLUE = 3

Color.RED.name, Color.RED.value
```

### 2.3. Nullable Types

```python
from typing import Optional, Union

# Optional[str] = str | None (Python 3.10+)
def greet(name: Optional[str] = None) -> str:
    if name is None:
        return "Hello, World!"
    return f"Hello, {name}!"

greet()
greet("John")
```

### 2.5. Generics

```python
from typing import TypeVar, Generic, List, Dict

T = TypeVar('T')

class Container(Generic[T]):
    def __init__(self, value: T):
        self.value = value
    def get(self) -> T:
        return self.value

int_container = Container(42)
str_container = Container("hello")
int_container.get()
```

### 2.6. Collection Operations

```python
numbers = [1, 2, 3, 4, 5]

# Map
squared = list(map(lambda x: x ** 2, numbers))

# Filter
evens = list(filter(lambda x: x % 2 == 0, numbers))

# Reduce
from functools import reduce
total = reduce(lambda acc, x: acc + x, numbers, 0)

# Sort
sorted_nums = sorted(numbers, reverse=True)

# Find first
result = next((x for x in numbers if x > 3), None)

# Zip
names = ["Alice", "Bob"]
ages = [25, 30]
combined = list(zip(names, ages))

print(f"squared: {squared}")
print(f"evens: {evens}")
print(f"total: {total}")
print(f"result: {result}")
print(f"combined: {combined}")
```

## 3. Tầng 3: OOP & Type Relationships

### 3.1. Class/Object

```python
class User:
    species = "Human"  # Class attribute

    def __init__(self, name: str, age: int):
        self.name = name
        self._age = age  # Protected (convention)

    def greet(self) -> str:
        return f"Hello, I'm {self.name}"

user = User("John", 30)
user.greet()
```

```python
# Data Class (Python 3.7+)
from dataclasses import dataclass

@dataclass
class Point:
    x: int
    y: int

p = Point(10, 20)
p.x, p.y
```

### 3.2. Kế Thừa & Đa Hình

```python
class Animal:
    def __init__(self, name: str):
        self.name = name
    def speak(self) -> str:
        return "..."

class Dog(Animal):
    def speak(self) -> str:
        return "Woof!"

class Cat(Animal):
    def speak(self) -> str:
        return "Meow!"

dog = Dog("Buddy")
dog.speak()
```

### 3.3. Interface/Trait/Protocol

```python
from typing import Protocol

class Drawable(Protocol):
    def draw(self) -> None: ...

class Circle:
    def draw(self) -> None:
        print("Drawing circle")

def render(shape: Drawable) -> None:
    shape.draw()

render(Circle())
```

## 4. Tầng 4: Memory Model

```python
# Garbage Collection
import gc

a = object()
b = a
del a
del b

# Force GC
gc.collect()

# Weak Reference
import weakref

class MyClass:
    def __init__(self, name):
        self.name = name

obj = MyClass("test")
ref = weakref.ref(obj)
ref() is obj  # True
```

### 4.2. Property & Getter/Setter

```python
class Temperature:
    def __init__(self):
        self._celsius = 0

    @property
    def celsius(self):
        return self._celsius

    @celsius.setter
    def celsius(self, value):
        self._celsius = value

    @property
    def fahrenheit(self):
        return self._celsius * 9/5 + 32

temp = Temperature()
temp.celsius = 25
temp.fahrenheit
```

## 5. Tầng 5: Concurrency Model

### 5.1. Concurrency/Async

```python
# Threading
import threading
import time

def worker(name, delay):
    print(f"{name} started")
    time.sleep(delay)
    print(f"{name} finished")

t1 = threading.Thread(target=worker, args=("Thread-1", 1))
t2 = threading.Thread(target=worker, args=("Thread-2", 1))
t1.start()
t2.start()
t1.join()
t2.join()
print("Done")
```

```python
# Async/Await
import asyncio

async def fetch_data():
    print("Fetching data...")
    await asyncio.sleep(1)
    return {"data": "result"}

async def main():
    result = await fetch_data()
    print(result)

asyncio.run(main())
```

## 6. Tầng 6: Paradigm

### 6.1. Functional Programming

```python
# Pure function
def add(a, b):
    return a + b

# Immutability - dùng tuple/frozen dataclass
from dataclasses import dataclass

@dataclass(frozen=True)
class Point:
    x: int
    y: int

# Function composition
from functools import reduce

def compose(*functions):
    return reduce(lambda f, g: lambda x: f(g(x)), functions)

f = compose(lambda x: x + 1, lambda x: x * 2)
f(3)  # (3 * 2) + 1 = 7
```

## 12. Tầng 12: Advanced Concepts

### 12.1. Advanced Concepts

```python
# Reflection
type(42)
isinstance(42, int)

# Introspection
import inspect

def func(a: int, b: str = "default") -> bool:
    pass

sig = inspect.signature(func)
list(sig.parameters.items())
```

```python
# Decorator
import functools
import time

def timer(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"{func.__name__} took {end - start:.4f}s")
        return result
    return wrapper

@timer
def slow_function():
    time.sleep(0.5)

slow_function()
```

```python
# Context Manager
class Timer:
    def __enter__(self):
        self.start = time.time()
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        self.end = time.time()
        print(f"Elapsed: {self.end - self.start}")

with Timer() as timer:
    time.sleep(0.3)

# contextlib
from contextlib import contextmanager

@contextmanager
def managed_resource():
    print("Acquiring")
    yield "resource"
    print("Releasing")

with managed_resource() as r:
    print(f"Using {r}")
```
